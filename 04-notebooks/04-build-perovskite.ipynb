```python
from perovskite_toolkit import PerovskiteStructureBuilder
```

```python
# obtain the conformer and modify the geometry manually
for identifier in identifier_list:
    perovskite_structure_builder = PerovskiteStructureBuilder(identifier=identifier)
    perovskite_structure_builder.generate_conformer(write_xyz_file=True)

# Use avocadro to adjust the torsion angle. benzene-benzene 20 degrees, benzene-thiophene 0 degree, thiophene-thiophene 0 degree
```

```python
identifier_list = [35490]
```

```python
# obtain organic molecule dimension and construct perovskite structure 
for identifier in identifier_list:
    perovskite_structure_builder = PerovskiteStructureBuilder(identifier=identifier)
    perovskite_structure_builder.find_calibration_vector()
    perovskite_structure_builder.calibrate_molecule(save_dimension=True)
    perovskite_structure_builder.build_inorganic_framework()
    perovskite_structure_builder.build_perovskite(save_cif=True)
    #perovskite_structure_builder.mol_dimension
    #perovskite_structure_builder.get_space_group_info()
```
